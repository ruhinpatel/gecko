## Converting data into format for the beta 3D application

Fields are:
ijk,Afreq,Bfreq,Cfreq,molecule,Beta,basis,omega 

where molecule is just the formula, and omega is a frequency shorthand where since we care about shg, we take the sum -(omegaB+omegaC)=omegaA and only take values where omegaB=omegaC.  So basically, we filter frequencies from omegaB and omegaC to keep only the cases where omegaB=omegaC and number the unique pairs starting with 0,0,0=1 (-(omega1+omega1),omega1,omega1)=2 ...

```python
shg_df=df.copy()
shg_df=shg_df[["ijk","omegaA","omegaB","omegaC","mol_id","value",'basis']]
renames={"value":"Beta",'omegaA':'Afreq','omegaB':'Bfreq','omegaC':'Cfreq','mol_id':'molecule'}
shg_df.rename(columns=renames, inplace=True)

# for every molecule, filter rows where omegaB==omegaC, and create a new column 'omega' that numbers the unique pairs starting from 1

shg_df['omega'] = None
for mol in shg_df['molecule'].unique():
    mol_df = shg_df[shg_df['molecule'] == mol]
    filtered = mol_df[mol_df['Bfreq'] == mol_df['Cfreq']]
    unique_freqs = filtered['Bfreq'].unique()
    freq_mapping = {freq: idx + 0 for idx, freq in enumerate(sorted(unique_freqs))}
    shg_df.loc[filtered.index, 'omega'] = filtered['Bfreq'].map(freq_mapping)


shg_df = shg_df.dropna(subset=["omega"])
```
